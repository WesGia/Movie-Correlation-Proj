```python
import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure
from importlib import reload

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

pd.options.mode.chained_assignment = None
```

```python
df = pd.read_csv('movies.csv')
```

```python
df.head()
```

## Cleaning Data

```python
df.isna().values.sum()
```

```python
df.duplicated().any()
```

```python
df.dropna(inplace=True)
```

```python
df.dtypes
```

```python
df['budget']= df['budget'].astype('int64')
df['gross'] = df['gross'].astype('int64')
```

```python
df.head()
```

```python
df['yearcorrect']=df['released'].str.extract(pat = '([0-9]{4})').astype(int)
df.head()
```

```python
df.sort_values(by=['gross'], inplace=False, ascending=False)
```

# **Scatterplot** 

budget vs gross

```python

plt.scatter(x=df['budget'], y=df['gross'])
plt.title('Budget vs Gross Earnings')
plt.xlabel('Budget')
plt.ylabel('Gross Earnings')
plt.show()
```

## Regression Plot


```python
sns.regplot(x='budget', y='gross', data=df,
            scatter_kws={'color':'red'}, line_kws={'color':'blue'})
```

```python
df.corr(method='pearson')
```

```python
correlation_matrix = df.corr(method='pearson')
sns.heatmap(correlation_matrix, annot=True)
plt.title('Correlation Matrix')
plt.xlabel('Movie Stats')
plt.ylabel('Movie Stats')
plt.show()
```

```python

df_numerized = df
for col in df_numerized.columns:
  if(df_numerized[col].dtype == 'object'):
    df_numerized[col]=df_numerized[col].astype('category')
    df_numerized[col]=df_numerized[col].cat.codes

```

```python
correlation_matrix = df_numerized.corr(method='pearson')
sns.heatmap(correlation_matrix, annot=True)
plt.title('Correlation Matrix')
plt.xlabel('Movie Stats')
plt.ylabel('Movie Stats')
plt.show()
```

```python
corr_mat = df_numerized.corr()
corr_pair = corr_mat.unstack()
sorted_pair = corr_pair.sort_values()
high_corr = sorted_pair[(sorted_pair)> 0.5]
high_corr
```

Votes and Budget have highest correlation to the gross earnings. Factors such as producer company, rating, and actors have low correlation
